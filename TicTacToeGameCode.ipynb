```python
print("<< Hello, Welcome to Tic Tac Toe Game >>") 
while(1):
    P1C = input("Please enter an X or O Character for Player # 1: ")
    # Capitalization...
    if (P1C != 'X' and P1C != 'O'):
        print("Invalid Input Character 'X' or 'O': ", P1C)
        continue
    if  P1C == 'X':
        P2C =  'O'
    else:
        P2C =  'X'
    break
print ('Player 1 Coice: ',P1C)
print ('Player 2 Coice: ',P2C)
print ('*** Proper Slot Ids:')
TTT1 = ['A1','A2','A3']
TTT2 = ['B1','B2','B3']
TTT3 = ['C1','C2','C3']
print(TTT1)
print(TTT2)
print(TTT3)
print('\n') # New Line!!!

# define list "TTT" for inputs
#TTT = ['-'*9]
#TTT = range(9) A numeric list ranging from 0 to 8
TTT1 = ['-','-','-','-']
TTT2 = ['-','-','-','-','-']
TTT = TTT1 + TTT2

def printTTT():
    print(TTT[0],TTT[1],TTT[2])
    print(TTT[3],TTT[4],TTT[5])
    print(TTT[6],TTT[7],TTT[8])
    return

print('TicTacToe Matrix Initialization:')
printTTT()

# print('...Test End...')

# Play the game...

# Define the function if_Available...

def if_Available(Pos,Chr):
    if  Pos == 'A1':
        if  TTT[0] == '-':
            TTT[0] = Chr
            return True  # empty cell
        else:
            return False  # occupied cell
    elif Pos == 'A2':
        if  TTT[1] == '-':
            TTT[1] = Chr
            return True   # empty cell
        else:
            return False  # occupied cell
    elif Pos == 'A3':
        if  TTT[2] == '-':
            TTT[2] = Chr
            return True  # empty cell
        else:
            return False  # occupied cell
    elif Pos == 'B1':
        if  TTT[3] == '-':
            TTT[3] = Chr
            return True  # empty cell
        else:
            return False  # occupied cell
    elif Pos == 'B2':
        if  TTT[4] == '-':
            TTT[4] = Chr
            return True  # empty cell
        else:
            return False  # occupied cell
    elif Pos == 'B3':
        if  TTT[5] == '-':
            TTT[5] = Chr
            return True   # empty cell
        else:
            return False  # occupied cell
    elif Pos == 'C1':
        if  TTT[6] == '-':
            TTT[6] = Chr
            return True   # empty cell
        else:
            return False  # occupied cell
    elif Pos == 'C2':
        if  TTT[7] == '-':
            TTT[7] = Chr
            return True   # empty cell
        else:
            return False  # occupied cell
    elif Pos == 'C3':
        if  TTT[8] == '-':
            TTT[8] = Chr
            return True   # empty cell
        else:
            return False  # occupied cell
    else:
            return False  # Wrong Slot Id
         
# Define the function Check_Draw() ...

def Check_Draw():
    # global ListLength
    ListLength = len(TTT)
    # print (ListLength)
    
    for i in range(ListLength):
          if  TTT[i] == '-':
            printTTT()
            return False
    printTTT()
    return True

# Define the function Check_Win() ...

def Check_Win():
    # Row A
    if  (TTT[0] != '-' and TTT[1] != '-' and TTT[2] != '-' and
         TTT[0] == TTT[1] and TTT[1] == TTT[2]
        ):
        return True
    # Row B
    if  (TTT[3] != '-' and TTT[4] != '-' and TTT[5] != '-' and
         TTT[3] == TTT[4] and TTT[4] == TTT[5]
        ):
        return True
    # Row C
    if  (TTT[6] != '-' and TTT[7] != '-' and TTT[8] != '-' and
         TTT[6] == TTT[7] and TTT[7] == TTT[8]
        ):
        return True
    # Colume # 1
    if  (TTT[0] != '-' and TTT[3] != '-' and TTT[6] != '-' and
         TTT[0] == TTT[3] and TTT[3] == TTT[6]
        ):
        return True
    # Colume # 2
    if  (TTT[1] != '-' and TTT[4] != '-' and TTT[7] != '-' and
         TTT[1] == TTT[4] and TTT[4] == TTT[7]
        ):
        return True
    # Colume # 3
    if  (TTT[2] != '-' and TTT[5] != '-' and TTT[8] != '-' and
         TTT[2] == TTT[5] and TTT[5] == TTT[8]
        ):
        return True
     # Diagonal "\"
    if  (TTT[0] != '-' and TTT[4] != '-' and TTT[8] != '-' and
         TTT[0] == TTT[4] and TTT[4] == TTT[8]
        ):
        return True
    # Diagonal "/"
    if  (TTT[2] != '-' and TTT[4] != '-' and TTT[6] != '-' and
         TTT[2] == TTT[4] and TTT[4] == TTT[6]
        ):
        return True
    return False
    
#class BreakoutException: pass  # declare a label

global Loop_Ok
Loop_Ok = 'True'

# print('TEST #1',Loop_Ok)

while (Loop_Ok == 'True' or Loop_Ok == 'Pass'):
    # print('#2',Loop_OK)
    if Loop_Ok == 'True': # Turn on Player 1 
        P1S   = input("Please enter Slot # for player # 1: ")
        Avlbl = if_Available(P1S,P1C)
        if  Avlbl: # True
            CD = Check_Draw()
            if CD: # True
                print('...No Winner...')
                printTTT()
                break
            CW = Check_Win()
            if CW: # True
                print('>>> Congrats to Player # 1')
                printTTT()
                break
            # raise BreakoutException
            # print('I am here... #1, Loop_Ok =', Loop_Ok)
            Loop_Ok = 'Pass' # next move by other player???
            continue
        else:
            # print('I am here... #2, Loop_Ok =', Loop_Ok)
            continue # read again for Player 1
    # read player 2
    # except BreakoutException:   # branching from Player 1 input to Player 2 input...
    while True:
        P2S   = input("Please enter Slot # for player # 2: ")
        Avlbl = if_Available(P2S,P2C)
        if  Avlbl: # True
            CD = Check_Draw()
            if CD: # True
                print('...No Winner...')
                printTTT()
                Loop_Ok = 'False'
                break
            CW = Check_Win()
            if CW: # True
                print('>>> Congrats to Player # 2')
                printTTT()
                Loop_Ok = 'False'
                break
            # print('I am here... #3, Loop_Ok =', Loop_Ok)
            Loop_Ok = 'True'
            # print('I am here... #4, Loop_Ok =', Loop_Ok)
            break # next move by Player 1
        else:
            # print('I am here... #5, Loop_Ok =', Loop_Ok)
            continue # read again for Player 2
 
while True:
    PlayAgain = input("Please enter Yes or No for playing again: ")
    if (PlayAgain != 'Yes' and PlayAgain != 'No'):
        print("Invalid Input!!! ", PlayAgain)
        continue
    if  PlayAgain == 'Yes':
            pass # Play the game (You need to branch to Program Start)...
            break
    else:
            print('Game End') # End the game...
            break
```
